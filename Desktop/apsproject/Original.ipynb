```python
import plotly as py
import plotly.graph_objs as go
import ipywidgets as widgets
import numpy as np
from scipy import special

py.offline.init_notebook_mode(connected=True)
```

```python
x1 = [1000,10000,100000,1000000,10000000]
y1 = [2.85,30.174,317.224,3774.8,44237.7]
y2 = [3.19,31.763,363.4,4534.1,51074.4]
y1 = [k/1000 for k in y1]
y2 = [k/1000 for k in y2]
layout = go.Layout(
            title='<b>INSERTION OF RANDOM NODES</b>',
            yaxis=dict(
                title='<i>Time Taken(sec)</i>'
            ),
            xaxis=dict(
                title='<i>Number of nodes</i>',
                type='category'
            )
        )
trace1 = go.Scatter(
            x=x1,
            y=y1,
            mode='lines',
            name='Binary Search Tree'
        )
trace2 = go.Scatter(
            x=x1,
            y=y2,
            mode='lines',
            name='Splay Tree'
        )
fig = go.Figure(data=[trace1,trace2], layout=layout)
py.offline.iplot(fig)
```

```python
x1 = [1000,10000,100000,1000000,10000000]
yy1 = [ 6.75 , 7.5 , 10.75 , 11 , 12.25 ]
yy2 = [ 9.25 , 10.5 , 11.25 , 12 , 14 ]
yy3 = [ 6 , 7.25 , 10 , 10.5 , 12 ]
yy4 = [ 5 , 4.5 , 5 , 4.75 , 5.5 ]
y1 = [k/1000 for k in yy1]
y2 = [k/1000 for k in yy2]
y3 = [k/1000 for k in yy3]
y4 = [k/1000 for k in yy4]
layout = go.Layout(
            title='<b>SEARCH IN RANDOM BUILT TREE</b>',
            yaxis=dict(
                title='<i>Time Taken(sec)</i>'
            ),
            xaxis=dict(
                title='<i>Number of nodes</i>',
                type='category'
            )
        )
trace6 = go.Scatter(
            x=x1,
            y=y1,
            mode='lines',
            name='First search in BST'
        )
trace7 = go.Scatter(
            x=x1,
            y=y2,
            mode='lines',
            name='First search in Splay Tree'
        )
trace8 = go.Scatter(
            x=x1,
            y=y3,
            mode='lines',
            name='Second search in BST'
        )
trace9 = go.Scatter(
            x=x1,
            y=y4,
            mode='lines',
            name='Second search in Splay Tree'
        )
fig = go.Figure(data=[trace6,trace8,trace7,trace9], layout=layout)
py.offline.iplot(fig)
```
